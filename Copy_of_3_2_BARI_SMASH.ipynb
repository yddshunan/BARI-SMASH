Since this is a longitudinal dataset, we have each construction ecometric for each year from 2010-2020. We also have a couple different versions of each ecometrics. I'm going to subset this dataframe so we just have one version of each ecometric for the year 2020, where each variable represents the proportion of land parcels (the land underneath a building) in a tract that received a given type of permit.

```python
permits_20 = permits[['CT_ID_10','NEWCON_PP.2020', 'DEMO_PP.2020', 'ADD_PP.2020', 'RENO_PP.2020']]
permits_20.rename(columns={"NEWCON_PP.2020": "New Construction",
                           "DEMO_PP.2020": "Demolitions",
                          "ADD_PP.2020": "Additions",
                          "RENO_PP.2020": "Renovations"}, inplace=True)
permits_20
```
